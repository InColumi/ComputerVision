```python
from io import BytesIO
from skimage import io
from PIL import Image, ImageDraw
from skimage import img_as_float
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
image = cv2.imread('1.jpg')
gray =cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
mean_image = cv2.blur(image, (5,5))
plt.figure(figsize=(5,5))
plt.imshow(mean_image, cmap='gray')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (25, 25))
#gradient = cv2.morphologyEx(mean_image, cv2.MORPH_CLOSE, kernel, iterations = 1)
gradient = cv2.morphologyEx(mean_image, cv2.MORPH_GRADIENT, kernel, iterations = 1)
plt.figure(figsize=(5,5))
plt.imshow(gradient, cmap="gray")
```

```python
#gradient1 = cv2.cvtColor(gradient,cv2.COLOR_BGR2GRAY)

def transformCircle(r):
    ret, thresh = cv2.threshold(gray, r, 255,cv2.THRESH_BINARY_INV)
    plt.figure(figsize=(5,5))
    plt.imshow(thresh, cmap='gray')
    
interact(transformCircle, r=(0,80))
```

```python
gradient_gray = cv2.cvtColor(gradient,cv2.COLOR_BGR2GRAY)
#ret, thresh = cv2.threshold(gradient_gray, 68, 255,cv2.THRESH_BINARY_INV)
ret, thresh = cv2.threshold(gray, 76, 255,cv2.THRESH_BINARY_INV)
plt.figure(figsize=(5,5))
plt.imshow(thresh, cmap='gray')
```

```python
# Маркер, показывающий, где точно находится фон
sure_bg = cv2.dilate(thresh,kernel,iterations=1)
plt.figure(figsize=(5,5))
plt.imshow(sure_bg, cmap='gray')
```

```python
# Маркер, показывающий, где точно находятся объекты
erode = cv2.erode(thresh,kernel,iterations = 1)
plt.figure(figsize=(5,5))
plt.imshow(erode, cmap='gray')
```

```python
sure_fg = np.uint8(erode)
unknown = cv2.subtract(sure_bg,sure_fg)

plt.figure(figsize=(5,5))
plt.imshow(unknown, cmap='gray')
```

```python
ret, markers = cv2.connectedComponents(sure_fg)

markers = markers+1
markers[unknown==255] = 0

markers = cv2.watershed(gradient,markers)
gradient[markers == -1] = [255,255,255]
plt.figure(figsize=(5,5))
plt.imshow(gradient, cmap='gray')
```

```python
markers = cv2.watershed(image,markers)
image[markers == -1] = [255,255,255]
plt.figure(figsize=(5,5))
plt.imshow(image, cmap='gray')
```
