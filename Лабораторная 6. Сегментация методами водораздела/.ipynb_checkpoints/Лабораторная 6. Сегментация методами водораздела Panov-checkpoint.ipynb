# Лабораторная 6. Сегментация методами водораздела

Выполнить сегментацию как на образце, применяя методы водораздела. Исходное изображение слева вверху. Получить все четыре изображения. Подробное описание: Гонсалес р.10.5

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

### Функция вывода изображения

```python
def showImage(images, size = 20, countInRow=2, countInCol=2, colorMap = 'gray'):
    countImages = len(images)
    #size *= countImages
    pic_box = plt.figure(figsize=(size, size))
    for i, image in enumerate(images):
        pic_box.add_subplot(countInRow, countInCol, i+1)
        plt.imshow(image)
        plt.set_cmap(colorMap)
        plt.subplots_adjust(wspace=0.1, hspace=0.1)
        plt.axis('off')
    plt.show()
```

### Исходное изображение

```python
image_srs = cv2.imread('1.png')
```

### Преобразование в серый цвет

```python
image_srs_gray = gray = cv2.cvtColor(image_srs,cv2.COLOR_BGR2GRAY)
```

```python
showImage([image_srs, image_srs_gray])
```

```python
bluredSobel = cv2.morphologyEx(image_srs, cv2.MORPH_GRADIENT, kernel, iterations = 11)
b = cv2.blur(bluredSobel, (5, 5))
showImage([bluredSobel, b, cv2.cvtColor(b, cv2.COLOR_BGR2GRAY)])
```

```python
sobelx = cv2.Sobel(b, cv2.CV_64F, 1, 0)
sobely = cv2.Sobel(b, cv2.CV_64F, 0, 1)

sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely, alpha=255 / sobely.max())

sobelxy = cv2.addWeighted(sobelx, 1, sobely, 0.5, 0)

sobelxy_blur = cv2.GaussianBlur(sobelxy, (5,5),cv2.BORDER_DEFAULT)

showImage([sobelxy, sobelxy_blur])
```

```python
ret, thresh = cv2.threshold(image_srs_gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
print(ret)
```

```python
kernel = np.ones((3,3), np.uint8)
```

```python
def transformCircle(r, s):
    bluredSobel = cv2.morphologyEx(image_srs, cv2.MORPH_GRADIENT, kernel, iterations = r)
    b = cv2.blur(bluredSobel, (s, s))
    showImage([bluredSobel, b, cv2.cvtColor(b, cv2.COLOR_BGR2GRAY)])

interact(transformCircle, r=(1, 50), s=(1, 10))
```

```python
bluredSobel = cv2.morphologyEx(image_srs_gray, cv2.MORPH_GRADIENT, kernel, iterations = 3)

showImage([cv2.blur(bluredSobel, (5, 5))])
```

```python
def transformCircle(r):
    t1 = cv2.erode(thresh, kernel, iterations = r)
    t1 = cv2.dilate(t1, kernel, iterations = r)
    t2 = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations = r)
    showImage([t1, t2])

interact(transformCircle, r=(1, 50))
```

```python
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations = 20)
```

```python
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations = 25)

sure_bg = cv2.dilate(opening, kernel, iterations=7)

dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 0)

ret, sure_fg = cv2.threshold(dist_transform, 0.35*dist_transform.max(), 255, 0)

sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)

ret, markers = cv2.connectedComponents(sure_fg)

# Add one to all labels so that sure background is not 0, but 1
markers = markers + 1

# Now, mark the region of unknown with zero
markers[unknown==255] = 0

image_with_markers = image_srs.copy()
markers = cv2.watershed(image_with_markers, markers)
image_with_markers[markers == -1] = [0, 255, 0]

showImage([image_srs,opening, dist_transform, sure_bg, sure_fg, unknown, markers, image_with_markers], size=50, countInCol=4)
```
