```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def get_dft_shifted(img):
    dft = cv.dft(np.float32(img), flags = cv.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    return dft_shift
```

```python
def get_spectrum(dft_shift):
    return cv.magnitude(dft_shift[:,:,0], dft_shift[:,:,1])
```

```python
def imshow(img):
    plt.imshow(img, cmap='gray')
    plt.show()
```

```python
def dist(u, v, m, n):
    return math.sqrt((u - m) * (u - m) + (v - n) * (v - n))
```

```python
img = cv.imread('13.png', cv.COLOR_BGR2GRAY)
img = img[:,:,2]
m,n = img.shape
print(img.shape)
plt.imshow(img, cmap = 'gray')
```

```python
dft_shift = get_dft_shifted(img)
spectrum = get_spectrum(dft_shift)
imshow(np.log(spectrum))
```

```python
# Идеальный фильтр
mask_I = np.zeros((m,n,2))
d = 30

for i in range(m):
    for j in range(n):
        mask_I[i][j][0] = int(dist(i, j, m/2, n/2) <= d)
        mask_I[i][j][1] = int(dist(i, j, m/2, n/2) <= d)
        
filtrated_I = dft_shift * mask_I

imshow(cv.magnitude(mask_I[:,:,0], mask_I[:,:,1]))
```

```python
idft = cv.dft(filtrated_I, cv.DFT_INVERSE)

idft_real = idft[:,:, 0]
idft_imaginary = idft[:,:, 1]

idft_magnitude = cv.magnitude(idft_real, idft_imaginary)
plt.imshow(idft, cmap = 'gray')
```

```python
mask_I_rev = 1 - mask_I
imshow(cv.magnitude(mask_I_rev[:,:,0], mask_I_rev[:,:,1]))
filtrated_I_rev = dft_shift * mask_I_rev
```

```python
idft = cv.dft(filtrated_I_rev, cv.DFT_INVERSE)

idft_real = idft[:,:,0]
idft_imaginary = idft[:,:,1]

idft = cv.magnitude(idft_real, idft_imaginary)
plt.imshow(idft, cmap='gray')
```

```python
dft_shift_I_rev = get_dft_shifted(idft)
spectrum_I_rev = get_spectrum(dft_shift_I_rev)
imshow(np.log(spectrum_I_rev))
imshow(np.log(spectrum))
```

```python
# Батерворт
print(dft_shift.shape)

mask_B = np.zeros((m, n ,2))
d = 30 
n_pow = 2

for i in range(m):
    for j in range(n):
        mask_B[i][j][0] = 1 / (1 + (dist(i, j, m/2, n/2) / d) ** (2*n_pow))
        mask_B[i][j][1] = 1 / (1 + (dist(i, j, m/2, n/2) / d) ** (2*n_pow))
        
filtrated_B = dft_shift * mask_B

imshow(cv.magnitude(mask_B[:,:,0], mask_B[:,:,1]))
```

```python
idft = cv.dft(filtrated_B, cv.DFT_INVERSE)

idft_real = idft[:,:, 0]
idft_imaginary = idft[:,:, 1]

idft_magnitude = cv.magnitude(idft_real, idft_imaginary)
plt.imshow(idft, cmap = 'gray')
```

```python
# не так отрабатывает
dft_shift_B = get_dft_shifted(idft)
spectrum_B = get_spectrum(dft_shift_B)
imshow(np.log(spectrum_B))
```

```python
mask_B_rev = np.zeros((m, n, 2))
d = 15
n_pow = 2

for i in range(m):
    for j in range(n):
        mask_B_rev[i][j][0] = 1 / (1 + (d / dist(i, j, m/2, n/2) / d) ** (2*n_pow))
        mask_B_rev[i][j][1] = 1 / (1 + (d / dist(i, j, m/2, n/2) / d) ** (2*n_pow))
        
filtrated_B_rev = dft_shift * mask_B_rev

imshow(cv.magnitude(mask_B_rev[:,:,0], mask_B_rev[:,:,1]))
```

```python
idft = cv.dft(filtrated_B_rev, cv.DFT_INVERSE)

idft_real = idft[:,:, 0]
idft_imaginary = idft[:,:, 1]

idft_magnitude = cv.magnitude(idft_real, idft_imaginary)
plt.imshow(idft, cmap = 'gray')
```

```python
# Гаусов
mask_G = np.zeros((m, n ,2))
d = 20 

for i in range(m):
    for j in range(n):
        mask_G[i][j][0] = math.exp((-1)*(dist(i, j, m/2, n/2) ** 2) / (2*d*d))
        mask_G[i][j][1] = math.exp((-1)*(dist(i, j, m/2, n/2) ** 2) / (2*d*d))
        
filtrated_G = dft_shift * mask_G

imshow(cv.magnitude(mask_G[:,:,0], mask_G[:,:,1]))
```

```python
idft = cv.dft(filtrated_G, cv.DFT_INVERSE)

idft_real = idft[:,:, 0]
idft_imaginary = idft[:,:, 1]

idft_magnitude = cv.magnitude(idft_real, idft_imaginary)
plt.imshow(idft, cmap = 'gray')
```

```python
# не так отрабатывает
dft_shift_G = get_dft_shifted(idft)
spectrum_G = get_spectrum(dft_shift_G)
imshow(np.log(spectrum_G + 0.001))
```

```python
mask_G_rev = 1 - mask_G
filtrated_G_rev = dft_shift * mask_G_rev
imshow(cv.magnitude(mask_G_rev[:,:,0], mask_G_rev[:,:,1]))
```

```python
idft = cv.dft(filtrated_G_rev, cv.DFT_INVERSE)

idft_real = idft[:,:,0]
idft_imaginary = idft[:,:,0]

idft = cv.magnitude(idft_real, idft_imaginary)
plt.imshow(idft, cmap='gray')
```

```python
dft_shift_G_rev = get_dft_shifted(idft)
spectrum_G_rev = get_spectrum(dft_shift_G_rev)
imshow(np.log(spectrum_G_rev))
```
