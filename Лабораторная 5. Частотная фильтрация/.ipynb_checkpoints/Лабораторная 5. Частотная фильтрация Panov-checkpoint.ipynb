```python
from io import BytesIO
from skimage import io
from PIL import Image, ImageDraw
from skimage import img_as_float
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 
import math
```

### Функция вывода изображения

```python
def showImage(images, size = 20, countInRow=2, countInCol=2, colorMap = 'gray'):
    countImages = len(images)
    #size *= countImages
    pic_box = plt.figure(figsize=(size, size))
    for i, image in enumerate(images):
        pic_box.add_subplot(countInRow, countInCol, i+1)
        plt.imshow(image)
        plt.set_cmap(colorMap)
        plt.subplots_adjust(wspace=0.1, hspace=0.1)
        plt.axis('off')
    plt.show()
```

### Функция для спектра изображения

```python
def get_spectrum(image: np.ndarray, const = 0):
    image_dft = cv2.dft(np.float32(image), flags = cv2.DFT_COMPLEX_OUTPUT)
    image_fftshift = np.fft.fftshift(image_dft)
    magnitude = cv2.magnitude(image_fftshift[:,:,0], image_fftshift[:,:,1])
    return np.log(magnitude + const)
```

### Функция нахождения расстояния между точками

```python
def get_distance(a, b, c, d):
    return math.sqrt((a - c) * (a - c) + (b - d) * (b - d))
```

# Функция перемножения спектра и маски

```python
def get_mull_spectrum_mask(spectrum, mask_perfect):
    width, height = spectrum.shape[0], spectrum.shape[1]
    result = np.zeros((width, height, 2))
    for i in range(width):
        for j in range(height):
            value = spectrum[i, j] * mask_perfect[i, j, 0]
            result[i, j, 0], result[i, j, 1] = value, value
    return result
```

# Функция обратного преобразования

```python
def get_processed_image(dft_shift, mask, constForSpectrum=0):
    filtered = dft_shift * mask
    shift_ifft = np.fft.ifftshift(filtered)
    idft = cv2.idft(shift_ifft)
    showImage([mask[:,:,0]])
    return cv2.magnitude(idft[:,:, 0], idft[:,:, 1])
```

# Функция вывод результата работы маски

```python
def show_processed_image(image, mask):
    dft = cv2.dft(np.float32(image), flags=cv2.DFT_COMPLEX_OUTPUT)
    image_dft_shift = np.fft.fftshift(dft)
    
    image_spectrum = get_spectrum(image)
    mull = get_mull_spectrum_mask(image_spectrum, mask)
    result_perfect = get_processed_image(image_dft_shift, mull)
    showImage([image,  result_perfect])
```

# Переменные для вычислений

# Идеальные фильтры низких частот (ИФНЧ)

###### Формула вычисления идеального фильтра

```python
mask_perfect_low = np.zeros((width, height, 2))
d = 30
for i in range(width):
    for j in range(height):
        distance = int(get_distance(i, j, width / 2, height / 2) <= d)
        mask_perfect_low[i][j] = (distance, distance)

show_processed_image(image_srs, mask_perfect_low)
```

# Идеальные фильтры высоких частот (ИФНЧ)

###### Формула вычисления идеального фильтра

```python
mask_perfect_high = np.zeros((width, height, 2))
d = 50
for i in range(width):
    for j in range(height):
        distance = int(get_distance(i, j, width / 2, height / 2) > d)
        mask_perfect_high[i][j] = (distance, distance)
        
show_processed_image(image_srs, mask_perfect_high)
```

# Фильтры низких частот Баттерворта (БФНЧ)

```python
mask_Butterworth_low = np.zeros((width, height ,2))
d = 30 
n = 2

for i in range(width):
    for j in range(height):
        value = 1.0 / (1 + (get_distance(i, j, width/2, height/2) / d) ** (2 * n))
        mask_Butterworth_low[i][j] = (value, value)

show_processed_image(image_srs, mask_Butterworth_low)
```

# Фильтры высоких частот Баттерворта (БФНЧ)

```python
mask_Butterworth_high = np.zeros((width, height ,2))
d = 30 
n = 2

for i in range(width):
    for j in range(height):
        distance = get_distance(i, j, width/2, height/2)
        value = 1.0 / (1 + (d / distance ) ** (2 * n))
        mask_Butterworth_high[i][j] = (value, value)
        
show_processed_image(image_srs, mask_Butterworth_high)
```

# Фильтры низких частот Гауса (ГФНЧ)

###### Формула вычисления низких частот Гауса

```python
mask_Gauss_low = np.zeros((width, height ,2))
d = 20 

for i in range(width):
    for j in range(height):
        distance = get_distance(i, j, width/2, height/2)
        value = math.exp(-(distance ** 2) / (2*d*d))
        mask_Gauss_low[i][j] = (value, value)
```

```python
show_processed_image(image_srs, mask_Gauss_low)
```

# Фильтры высоких частот Гауса (ГФНЧ)

```python
mask_Gauss_high =  1 - mask_Gauss_low
show_processed_image(image_srs, mask_Gauss_high)
```
