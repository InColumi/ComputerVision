```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline 
import numpy as np
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

# Функция вывода изображения

```python
def showImage(images, size = 15, colorMap = 'gray'):
    countImages = len(images)
    #size *= countImages
    pic_box = plt.figure(figsize=(size, size * countImages))
    for i, image in enumerate(images):
        pic_box.add_subplot(countImages, 2, i+1)
        plt.imshow(image)
        plt.set_cmap(colorMap)
        plt.subplots_adjust(wspace=0, hspace=0.1)
        plt.axis('off')
    plt.show()
```

# Загрузка исходного изображения

```python
imageSrs = cv2.imread("1.jpg")
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

# a) Полный снимок скелета, полученный с помощью гамма-лучей

```python
showImage([imageSrs])
```

# Преобразование исходного в серый цвет

```python
grayImage = cv2.cvtColor(imageSrs, cv2.COLOR_BGR2GRAY)
```

# б) Применение оператора лапласиана к изображению (а) (филтр с помощью ядра)

# Ядро для фильтра

```python
kernel1 = np.array([[-1, -1, -1],[-1, 8, -1],[-1, -1, -1]])
```

```python
filtered = cv2.filter2D(grayImage, 6, kernel1)
```

# в) Повышение резкости сложением изображений (а) и (б)

```python
sumGrayAndFiltered = grayImage + filtered
```

# г) Применение градиентного оператора Собела к изображению (а)

```python
sobelx = cv2.Sobel(grayImage, cv2.CV_64F, 1, 0)
sobely = cv2.Sobel(grayImage, cv2.CV_64F, 0, 1)

sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)

sobelxy = cv2.addWeighted(sobelx, 1, sobely, 1, 1)
```

# д) Изображение (г), сглаженное усредняющим фильтром по окрестности 5×5

```python
bluredSobel = cv2.blur(sobelxy, (5,5))
```

# е) Изображение-маска, полученное перемножением изображений (в) и (д)

```python
mask = sumGrayAndFiltered * bluredSobel
```

# ж) Изображение с повышенной резкостью, полученное сложением изображений (а) и (е)

```python
hightSharpness = grayImage + mask
```

# з) Конечный результат, полученный из изображения (ж) градационной коррекцией по степенному закону

```python
gamma = 0.5
hightSharpness.astype(float)

result = np.array(255*(hightSharpness / 255) ** gamma, dtype = 'uint8')
```

# Вывод изображений

```python
showImage([imageSrs, filtered, sumGrayAndFiltered, sobelxy, bluredSobel, mask, hightSharpness, result])
```

```python
def transformCircle(r):
    filtered = cv2.filter2D(grayImage, r, kernel1)
    showImage([filtered])

interact(transformCircle, r=(-1, 6))
```
